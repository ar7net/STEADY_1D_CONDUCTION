```python
from IPython.display import Markdown as md
import numpy as np
from numpy import linalg
import matplotlib as mpl
import matplotlib.pyplot as plt
from num2tex import num2tex
import pandas as pd

from IPython.core.magic import register_cell_magic
from IPython.display import HTML, display
my_palette = ['#edc6a3','#f4ce88','#f0e9ee','#cbcfd9','#c9dac7','#d1d3be','#efe1d5','#e6b4a7','#c0b6a2','#cdb090']
my_palette_names = {'getter':'#e6b4a7', 'init':'#efe1d5'} 
@register_cell_magic
def bgc(color, cell=None):
    if str.isdigit(color) and len(color) == 1:
        color = my_palette[int(color)]
    else:
        color = my_palette_names.get(color, color)
    script = ("var cell = this.closest('.code_cell');"
    "var editor = cell.querySelector('.input_area');"
    "editor.style.background='{}';"
    "this.parentNode.removeChild(this)").format(color) 
    display(HTML('<img src onerror="{}">'.format(script)))

#--------------------------------------BLOQUE DE CÓDIGO--------------------------------------------------------
s=1   # Sentido de giro: (1) horario, (2) antihorario
w_AB=4                                            # Velocidad angular manivela
alfa_AB=0                                          # Aceleración angular manivela
r_AB=np.array([140, -100,0])								# ¿Cuál es la posición de B con respecto de A?
r_BC=np.array([0, -200,0])		# ¿Cuál es la posición de C con respecto de B?
r_DC=np.array([300, -100,0])	# ¿Cuál es la posición de C con respecto de D?

#---------------------------------------------------------------------------------------------------------------
if s==1:

    # Manivela, movimiento de rotación
    w_AB_VEC=np.array([0,0,-w_AB])
    alfa_AB_VEC=np.array([0,0,-alfa_AB])
    v_B=np.cross(w_AB_VEC,r_AB)
    a_B=np.cross(alfa_AB_VEC, r_AB)-((np.linalg.norm(w_AB_VEC))**2)*r_AB

    # Biela, movimiento plano general
    A=np.array( [   [1,0,r_BC[1],0],[0,1,-r_BC[0],0],[1,0,0,r_DC[1]],[0,1,0,-r_DC[0]]     ]   )
    b=np.array([v_B[0],v_B[1],0,0])
    solu = np.linalg.solve(A,b)
    v_cx=solu[0]
    v_cy=solu[1]
    w_BC=solu[2]
    w_DC=solu[3]

    v_C=np.array([v_cx,v_cy,0])
    w_BC_VEC=np.array([0,0,w_BC])
    w_DC_VEC=np.array([0,0,w_DC])

    B=np.array([[1,0,r_BC[1],0], [0,1, -r_BC[0],0],  [1,0,0, r_DC[1]], [0,1,0,-r_DC[0]]])
    c=np.array([a_B[0]-((np.linalg.norm(w_BC_VEC))**2)*r_BC[0], a_B[1]-((np.linalg.norm(w_BC_VEC))**2)*r_BC[1], -((np.linalg.norm(w_DC_VEC))**2)*r_DC[0], -((np.linalg.norm(w_DC_VEC))**2)*r_DC[1]])
    solu1=np.linalg.solve(B,c)
    a_C=np.array([solu1[0], solu1[1],0])
    alfa_BC=np.array([0,0, solu1[2]])
    alfa_DC=np.array([0,0, solu1[3]])
    
    k_1=(w_BC_VEC[2]**2)*r_BC[1]
    k_2=(w_BC_VEC[2]**2)*r_BC[0]
        
    k_111=a_B[0]-k_1   # forma parcial
    k_222=a_B[1]-k_2   # forma parcial
        
    k_11=a_B[0]-alfa_BC[2]*r_BC[1]-k_1   # respuesta final de a_c (en direccion i)
    k_22=a_B[1]-alfa_BC[2]*r_BC[0]-k_2   # respuesta final  de a_c (en direccion j)
    alfa_BC_abs=abs(alfa_BC[2])
    alfa_DC_abs=abs(alfa_DC[2])
   
     

else:
    if s==2:
        # Manivela, movimiento de rotación
        w_AB_VEC=np.array([0,0, w_AB])
        alfa_AB_VEC=np.array([0,0, alfa_AB])
        v_B=np.cross(w_AB_VEC,r_AB)
        a_B=np.cross(alfa_AB_VEC, r_AB)-((np.linalg.norm(w_AB_VEC))**2)*r_AB

        # Biela, movimiento plano general
        A=np.array( [   [1,0,r_BC[1],0],[0,1,-r_BC[0],0],[1,0,0,r_DC[1]],[0,1,0,-r_DC[0]]     ]   )
        b=np.array([v_B[0],v_B[1],0,0])
        solu = np.linalg.solve(A,b)
        v_cx=solu[0]
        v_cy=solu[1]
        w_BC=solu[2]
        w_DC=solu[3]

        v_C=np.array([v_cx,v_cy,0])
        w_BC_VEC=np.array([0,0,w_BC])
        w_DC_VEC=np.array([0,0,w_DC])

        B=np.array([[1,0,r_BC[1],0], [0,1, -r_BC[0],0],  [1,0,0, r_DC[1]], [0,1,0,-r_DC[0]]])
        c=np.array([a_B[0]-((np.linalg.norm(w_BC_VEC))**2)*r_BC[0], a_B[1]-((np.linalg.norm(w_BC_VEC))**2)*r_BC[1], -((np.linalg.norm(w_DC_VEC))**2)*r_DC[0], -((np.linalg.norm(w_DC_VEC))**2)*r_DC[1]])
        solu1=np.linalg.solve(B,c)
        a_C=np.array([solu1[0], solu1[1],0])
        alfa_BC=np.array([0,0, solu1[2]])
        alfa_DC=np.array([0,0, solu1[3]])
        
        k_1=(w_BC_VEC[2]**2)*r_BC[1]
        k_2=(w_BC_VEC[2]**2)*r_BC[0]
        
        k_111=a_B[0]-k_1   # forma parcial
        k_222=a_B[1]-k_2   # forma parcial
        
        k_11=a_B[0]-alfa_BC[2]*r_BC[1]-k_1   # respuesta final de a_c (en direccion i)
        k_22=a_B[1]-alfa_BC[2]*r_BC[0]-k_2   # respuesta final  de a_c (en direccion j)
        alfa_BC_abs=abs(alfa_BC[2])
        alfa_DC_abs=abs(alfa_DC[2])
     
```
